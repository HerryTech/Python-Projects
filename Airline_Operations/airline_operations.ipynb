#### Why CANCELLATION REASON has missing values

```python
flights['CANCELLED'].value_counts()
```

- Because flights that were not cancelled should not have a cancellation reason.
- These missing values are meaningful.
- Therefore, can't be filled or removed.

#### Why AIR_SYSTEM_DELAY, SECURITY_DELAY, AIRLINE_DELAY, LATE_AIRCRAFT_DELAY, WEATHER_DELAY have missing values

- AIR_SYSTEM_DELAY, SECURITY_DELAY, AIRLINE_DELAY, LATE_AIRCRAFT_DELAY, WEATHER_DELAY are often only populated when a flight experienced a reportable delay.
- Most flights either arrived on time, arrived early or had minor delays
- Therefore no official delay cause was recorded

#### Why DEPARTURE_TIME, DEPARTURE_DELAY, TAXI_OUT, WHEELS_OFF, ARRIVAL_TIME, ARRIVAL_DELAY, AIR_TIME, ELAPSED_TIME have missing values

- These numbers are suspiciously close to CANCELLATION_REASON non-null count of 89,884

- This suggests that many of these missing values belong to cancelled flights.
- Which is completely logical. A cancelled flight never departed, taxied, landed or accumulated air time

###### Interpretation of Departure Delay Distribution

The departure delay distribution is heavily right skewed, meaning that most flights experienced little to no departure delay, while a relatively small number of flights experienced very large delays. This is evident from both the histogram and the summary statistics. The median departure delay is -1 minute, indicating that at least half of the flights departed on time or slightly early. The mean departure delay is 11.33 minutes, which is much higher than the median because a small number of extremely delayed flights pull the average upward. The minimum delay of -61 minutes shows that some flights departed significantly earlier than scheduled, while the maximum delay of 1,988 minutes demonstrates the presence of rare but severe operational disruptions. Overall, the distribution suggests that large departure delays are uncommon, but when they occur, they can be substantial enough to significantly affect the overall average.

##### Interpretation of Arrival Delay Distribution

The arrival delay distribution also exhibits a strong right skew. Most flights arrived close to schedule, with many arriving early or only slightly late. This can be seen from the median arrival delay of -3 minutes, indicating that more than half of flights arrived on time or earlier than scheduled. The mean arrival delay of 7.61 minutes is higher than the median due to the influence of a relatively small number of flights with very large delays. The minimum arrival delay of -82 minutes indicates that some flights arrived considerably earlier than scheduled, while the maximum delay of 1,971 minutes highlights the occurrence of rare but extreme disruptions. Similar to departure delays, the majority of flights cluster around zero delay, while a small number of severely delayed flights create a long right tail in the distribution.

```python
summary = pd.DataFrame({
    'Metric': [
        'Departure Delay Rate',
        'Arrival Delay Rate',
        'Cancellation Rate',
        'Diversion Rate'
    ],
    "Count": [
        (df['DEPARTURE_DELAY'] > 0).sum(),
        (df['ARRIVAL_DELAY'] > 0).sum(),
        (df['CANCELLED'] == 1).sum(),
        (df["DIVERTED"] == 1).sum()
    ],
    'Value (%)': [
        (df['DEPARTURE_DELAY'] > 0).mean() * 100,
        (df['ARRIVAL_DELAY'] > 0).mean() * 100,
        df['CANCELLED'].mean() * 100,
        df['DIVERTED'].mean() * 100
    ]
})

summary
```

#### Business Interpretation

- The airline's operational performance shows a substantial level of disruption. Approximately 40% of all flights experienced departure delays and slightly more than 40% arrived late. This means that roughly four out of every ten flights failed to meet their scheduled departure or arrival times. Such a high proportion of delayed flights is likely to have a significant impact on passenger satisfaction and supports management's concern regarding increasing customer complaints.

- Although cancellations account for only 3.86% of all flights, they represent over 40,000 cancelled flights during the year. From a customer experience perspective, cancellations are often more disruptive than delays because passengers may need to rebook, wait for alternative flights, or abandon travel plans altogether. Therefore, despite being less frequent than delays, cancellations should be considered a critical area of investigation.

- Diversions are relatively rare, affecting only 0.24% of flights. While diversions can create severe disruptions for affected passengers, their low occurrence suggests they are unlikely to be a primary driver of overall customer dissatisfaction across the network.

- Overall, the results indicate that delays, rather than cancellations or diversions, are the dominant operational issue and should be the primary focus of subsequent analysis.

```python
monthly_delay = (
    df.groupby('MONTH')
      .agg({
          'DEPARTURE_DELAY': 'mean',
          'ARRIVAL_DELAY': 'mean',
          'CANCELLED': 'mean'
      })
      .reset_index()
)

monthly_delay['CANCELLED'] = monthly_delay['CANCELLED'] * 100

monthly_delay
```

```python
import matplotlib.pyplot as plt

monthly_delay.plot(
    x='MONTH',
    y=['DEPARTURE_DELAY', 'ARRIVAL_DELAY'],
    figsize=(10,5),
    marker='o'
)

plt.title('Average Monthly Delays')
plt.ylabel('Minutes')
plt.show()
```

```python
monthly_delay.plot(
    x='MONTH',
    y='CANCELLED',
    figsize=(10,5),
    marker='o'
)

plt.title('Monthly Cancellation Rate')
plt.ylabel('Cancellation Rate (%)')
plt.show()
```

#### Interpretation of the Monthly Delay Trend

- The average departure and arrival delays increased steadily over the three month period. Average departure delays rose from approximately 9.76 minutes in January to 11.89 minutes in February, reaching 14.84 minutes in March. Similarly, average arrival delays increased from 5.81 minutes in January to 8.32 minutes in February and 11.40 minutes in March. This consistent upward trend suggests that flight punctuality deteriorated during the first quarter of the year, with passengers experiencing progressively longer delays each month.

- An interesting observation is that departure delays remain consistently higher than arrival delays across all three months. This suggests that although many flights departed behind schedule, airlines were able to recover part of the lost time while en route. Factors such as favorable weather, efficient flight operations, or adjustments in flight speed may have contributed to reducing arrival delays relative to departure delays.

#### Interpretation of the Monthly Cancellation Rate

- The monthly cancellation rate also increased throughout the three month period. Approximately 2.55% of flights were cancelled in January, rising to 4.78% in February and 5.37% in March. This indicates that operational disruptions became more frequent as the quarter progressed. The simultaneous increase in both delays and cancellations suggests that the airline system experienced increasing operational pressure during these months. Although further investigation is required to determine the exact causes, factors such as adverse weather conditions, increased traffic demand, or operational constraints may have contributed to the worsening performance.
