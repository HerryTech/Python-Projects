#### Why CANCELLATION REASON has missing values

```python
flights['CANCELLED'].value_counts()
```

- Because flights that were not cancelled should not have a cancellation reason.
- These missing values are meaningful.
- Therefore, can't be filled or removed.

#### Why AIR_SYSTEM_DELAY, SECURITY_DELAY, AIRLINE_DELAY, LATE_AIRCRAFT_DELAY, WEATHER_DELAY have missing values

- AIR_SYSTEM_DELAY, SECURITY_DELAY, AIRLINE_DELAY, LATE_AIRCRAFT_DELAY, WEATHER_DELAY are often only populated when a flight experienced a reportable delay.
- Most flights either arrived on time, arrived early or had minor delays
- Therefore no official delay cause was recorded

#### Why DEPARTURE_TIME, DEPARTURE_DELAY, TAXI_OUT, WHEELS_OFF, ARRIVAL_TIME, ARRIVAL_DELAY, AIR_TIME, ELAPSED_TIME have missing values

- These numbers are suspiciously close to CANCELLATION_REASON non-null count of 89,884

- This suggests that many of these missing values belong to cancelled flights.
- Which is completely logical. A cancelled flight never departed, taxied, landed or accumulated air time

This means that roughly one in every three flights experienced some level of delay, suggesting that schedule reliability remains an important operational challenge. Although the majority of flights operated on time or arrived early, delayed flights still represent a substantial portion of the network and are likely to contribute significantly to customer dissatisfaction.

- Cancellations accounted for 1.54% of all scheduled flights, representing 89,884 cancelled flights during the year. While cancellations occur far less frequently than delays, they often have a greater impact on passengers because they may result in missed connections, rebooking, or cancelled travel plans. Consequently, cancellations remain an important operational metric despite their relatively low frequency.

- Diversions affected only 0.26% of all flights, indicating that they are relatively uncommon events. Although diversions can significantly inconvenience affected passengers, their low occurrence suggests they are unlikely to be a primary driver of overall operational performance.

- Overall, flight delays emerge as the most significant operational issue in the dataset. Their frequency and scale suggest that improving schedule reliability should be a priority for reducing customer complaints and enhancing operational efficiency.

### Trend Analysis

#### Monthly Delay Trends

```python
monthly_delay = (
    df.groupby('MONTH')
      .agg({
          'DEPARTURE_DELAY': 'mean',
          'ARRIVAL_DELAY': 'mean',
          'CANCELLED': 'mean'
      })
      .reset_index()
)

monthly_delay['CANCELLED'] = monthly_delay['CANCELLED'] * 100

monthly_delay
```

```python
import matplotlib.pyplot as plt

monthly_delay.plot(
    x='MONTH',
    y=['DEPARTURE_DELAY', 'ARRIVAL_DELAY'],
    figsize=(10,5),
    marker='o'
)

plt.title('Average Monthly Delays')
plt.ylabel('Minutes')
plt.show()
```

```python
monthly_delay.plot(
    x='MONTH',
    y='CANCELLED',
    figsize=(10,5),
    marker='o'
)

plt.title('Monthly Cancellation Rate')
plt.ylabel('Cancellation Rate (%)')
plt.show()
```

#### Interpretation of the Monthly Delay Trend

- The monthly analysis reveals that flight delays do not follow a steady upward or downward trend throughout the year. Instead, operational performance fluctuates across different months, suggesting the influence of seasonal factors. Average departure and arrival delays were relatively high during January, February, June, July, and December, while September and October recorded the lowest average delays of the year. Notably, the average arrival delay became slightly negative during September and October, indicating that flights, on average, arrived earlier than their scheduled arrival times during these months.

- Another notable observation is that departure delays remain consistently higher than arrival delays across every month. This suggests that airlines are often able to recover part of the lost time during flight operations, resulting in better arrival performance than departure performance. Overall, the results indicate that flight punctuality varies throughout the year and may be influenced by seasonal travel demand, weather conditions, and operational pressures.

#### Interpretation of the Monthly Cancellation Rate

- Cancellation rates also display clear seasonal variation rather than a consistent trend. The highest cancellation rates occurred in February (4.78%) and January (2.55%), while the lowest rates were observed during September (0.45%) and October (0.50%). This pattern closely aligns with the delay trends, suggesting that the factors contributing to increased delays during the winter months may also lead to a greater number of cancelled flights. The noticeable reduction in cancellations during early autumn indicates a period of relatively stable flight operations.

#### Are delays consistent throughout the week, or are certain days performing worse than others?

#### Average Delay by Day of Week

```python
weekday_delay = (
    df.groupby('DAY_OF_WEEK')
      .agg({
          'DEPARTURE_DELAY': 'mean',
          'ARRIVAL_DELAY': 'mean',
          'CANCELLED': 'mean'
      })
      .reset_index()
)

weekday_delay['CANCELLED'] *= 100

weekday_delay
```

```python
day_map = {
    1: 'Monday',
    2: 'Tuesday',
    3: 'Wednesday',
    4: 'Thursday',
    5: 'Friday',
    6: 'Saturday',
    7: 'Sunday'
}

weekday_delay['DAY_NAME'] = weekday_delay['DAY_OF_WEEK'].map(day_map)
```

```python
weekday_delay
```

```python
import matplotlib.pyplot as plt

weekday_delay.plot(
    x='DAY_NAME',
    y=['DEPARTURE_DELAY', 'ARRIVAL_DELAY'],
    figsize=(10,5),
    marker='o'
)

plt.title('Average Delay by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Average Delay (Minutes)')
plt.xticks(rotation=45)
plt.show()
```

```python
weekday_delay.plot(
    x='DAY_NAME',
    y='CANCELLED',
    figsize=(10,5),
    marker='o'
)

plt.title('Cancellation Rate by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Cancellation Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

#### Interpretation of Average Delay by Day of Week

- The analysis shows that flight delays vary throughout the week, although the differences are not extremely large. Monday recorded the highest average departure delay of 10.87 minutes and the highest average arrival delay of 6.00 minutes, making it the poorest performing day in terms of schedule reliability. In contrast, Saturday experienced the lowest average departure delay of 7.83 minutes and the lowest average arrival delay of 1.85 minutes, indicating that flights generally operated more punctually on weekends, particularly on Saturdays.

- Across all seven days, average departure delays consistently exceed average arrival delays. This suggests that airlines are able to recover some of the time lost before takeoff during the flight, allowing many delayed departures to arrive closer to schedule. The relatively small variation between weekdays also indicates that delays are not isolated to a single day but are a recurring operational challenge throughout the week.

#### Interpretation of Cancellation Rate by Day of Week

- The cancellation analysis also reveals noticeable differences across the week. Monday has the highest cancellation rate at 2.43%, followed by Tuesday at 1.78% and Sunday at 1.61%. Friday recorded the lowest cancellation rate at 1.02%, while Saturday and Wednesday also maintained relatively low cancellation rates.

- The fact that Monday experiences both the highest average delays and the highest cancellation rate suggests that operational challenges are greatest at the beginning of the week. One possible explanation is that disruptions from the weekend may carry over into Monday operations, affecting aircraft availability, crew scheduling, and flight schedules. Conversely, Saturday's strong performance may be associated with lower business travel demand and reduced airport congestion, allowing airlines to operate more efficiently

- This analysis suggests that operational performance is influenced by the day of the week. Although delays occur throughout the week, the beginning of the week appears to present greater operational challenges, while Saturday demonstrates the strongest on time performance. These findings indicate that airline managers may benefit from reviewing staffing levels, aircraft scheduling, and operational planning for Mondays to improve service reliability and reduce passenger disruption.
